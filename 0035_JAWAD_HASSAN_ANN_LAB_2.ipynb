```python
#libraries
from tensorflow import keras
import numpy as np
import tensorflow as tf
```

```python
#lab activity:

#dataset
x=np.array([[0],[1],[2],[3],[4],[5]],dtype=float)
y=np.array([[0],[0],[0],[1],[1],[1]],dtype=float)

#model network
model=keras.Sequential([
 keras.layers.Dense(6,activation='relu',input_shape=(1,)),
 keras.layers.Dense(1,activation='sigmoid')
])


#model compile
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

#model train
model.fit(x,y,epochs=100,verbose=1)

#model predict
prediction=model.predict(np.array([[4]]))
print(prediction)
```

```python
#task1 
# #using sigmoid in hidden layer decreased the accuracy from 66 to 50 percent because sigmoid can cause vanishing gradient
# problem which makes it harder for the model to learn and converge.
# ReLU is a better choice for hidden layers as it allows for faster training and better performance.
x=np.array([[0],[1],[2],[3],[4],[5]],dtype=float)
y=np.array([[0],[0],[0],[1],[1],[1]],dtype=float)
model=keras.Sequential([
 keras.layers.Dense(6,activation='sigmoid',input_shape=(1,)),
 keras.layers.Dense(1,activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model.fit(x,y,epochs=100,verbose=1)
model.predict(np.array([[4]]))
```

```python
#task2
x=np.array([[0],[1],[2],[3],[4],[5]],dtype=float)
y=np.array([[0],[0],[0],[1],[1],[1]],dtype=float)
model=keras.Sequential([
 keras.layers.Dense(6,activation='tanh',input_shape=(1,)),
 keras.layers.Dense(1,activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model.fit(x,y,epochs=100,verbose=1)
p=model.predict(np.array([[4]]))
print('prediction:',p)
#using tanh instead of relu in hidden layer for trainning the model, tanh gave 81 percent prediction while relu gave 
#85 percent, which is better because relu is less likely to cause vanishing gradient problem.
```

```python
#task3
#10 neurons in hidden layer
x=np.array([[0],[1],[2],[3],[4],[5]],dtype=float)
y=np.array([[0],[0],[0],[1],[1],[1]],dtype=float)
model=keras.Sequential([
    keras.layers.Dense(10,activation='relu',input_shape=(1,)),
    keras.layers.Dense(1,activation='sigmoid')  
])
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model.fit(x,y,epochs=100,verbose=1)
p=model.predict(np.array([[4]]))
print('prediction:',p)
```

```python

#6 neurons in hidden layer
model=keras.Sequential([
    keras.layers.Dense(6,activation='relu',input_shape=(1,)),
    keras.layers.Dense(1,activation='sigmoid')  
])
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model.fit(x,y,epochs=100,verbose=1)
p=model.predict(np.array([[4]]))
print('prediction:',p)
```

## Does accuracy improve?
### yes accuracy improved with 10 neuron but model overfit that is why prediction is 72 percent meanwhile with 6 neurons in hidden layer  prediction was 88 percent, which is better than 72 percent because increasing the number of neurons in the hidden layer can lead to overfitting, which can decrease the model's ability to generalize to new data and can lead to lower accuracy on unseen data.
## Does training become faster or slower?
### training becomes slower because increasing the number of neurons in the hidden layer increases the number of parameters in the model, which can lead to longer training times and increased computational resources required to train the model.

```python
#leaky relu in hidden layer:
model=keras.Sequential([
     keras.layers.Dense(6,input_shape=(1,)),
     keras.layers.LeakyReLU(alpha=0.05),
     keras.layers.Dense(1,activation='sigmoid')                   
])
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model.fit(x,y,epochs=100,verbose=1)
p=model.predict(np.array([[4]]))
print('prediction:',p)
```
