```python
#tic count
import time
import numpy as np
import matplotlib.pyplot as plt
start=time.perf_counter()
x=np.random.randint(0,1001,1000)
end=time.perf_counter()
print('time count:',end-start)
start=time.perf_counter()
x=np.sort(x)
end=time.perf_counter()
print('time count:',end-start)
start=time.perf_counter()
xsq=np.square(x)
end=time.perf_counter()
print('time count:',end-start)
start=time.perf_counter()
plt.plot(x,xsq)
plt.show()
end=time.perf_counter()
print('time count:',end-start)
```

```python
l=[7,3,9,1,2,8,6,-1]
for i in range(0,7):
    for j in range(i+1,8):
        if l[i]>l[j]:
            temp=l[i]
            l[i]=l[j]
            l[j]=temp
            
print(l)
        
```

```python
def hanoi(n,source,helper,destination):
    if n==1:
        print(source,'-->',destination)
    else:
        hanoi(n-1,source,destination,helper)
        print(source,'-->',destination)
        hanoi(n-1,helper,source,destination)
hanoi(4,'a','b','c')
```
