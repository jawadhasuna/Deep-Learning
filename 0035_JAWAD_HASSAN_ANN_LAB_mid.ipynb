## Question:01    вҖңScenario: The Smart Security "Dual-Sensor" AlarmвҖқ 

```python
#libraries:
import numpy as np
from sklearn.linear_model import Perceptron
```

```python
#q1

X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 1, 1, 1])
model=Perceptron()

#  Train the model
model.fit(X, y)

test = np.array([[0, 0]])
prediction = model.predict(test)

print("Smart Security System Test")
print(f"Testing Sensors at: {test}")
print(f"Model Prediction (Alarm State): {prediction}")

if prediction == 0:
    print("Result: System remains SILENT. No false positive detected.")
else:
    print("Result: ALARM triggered.")
```

### Question:03    ANN Model for Flower Classification

```python
#q3
#1 mount
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
#2 load
df=pd.read_csv('/Users/macbook/Desktop/sem6/ANN/mid/Iris - Iris.csv')
#3 separate
X = df.iloc[:, 1:5].values 
y = df.iloc[:, -1].values
#4 convert string labels to integers (Label Encoding) 
# and then to a binary matrix (One-Hot Encoding) for the Softmax output layer.
le = LabelEncoder()
integer = le.fit_transform(y)

# One-Hot Encoding: Convert 0 to [1, 0, 0], 1 to [0, 1, 0], and 2 to [0, 0, 1].
onehot_encoder = OneHotEncoder(sparse_output=False)
integer_encoded = integer.reshape(len(integer), 1)
y_encoded = onehot_encoder.fit_transform(integer_encoded)

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
#5 architecture
model = Sequential()
model.add(Dense(16, activation='relu', input_shape=(4,)))
model.add(Dense(8, activation='relu'))
model.add(Dense(3, activation='softmax'))
#6 compile
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
history = model.fit(X_train, y_train, epochs=50, batch_size=8, validation_data=(X_test, y_test))


scores = model.evaluate(X, y_encoded)
print(f"\nFinal Accuracy: {scores[1]*100:.2f}%")
#7 summary
model.summary()
```
